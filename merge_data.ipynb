```python
import pandas as pd
import glob
import os
```

找出所有 .dta 結尾的中國工業資料

```python
directory = 'C:\\Users\\peipeipig\\OneDrive\\桌面\\作品集\\中國政策分析\\DATA\\rawdata'

# dta_files = glob.glob(directory + '*.dta')
dta_files = [file for file in os.listdir(directory)]
```

合併所有資料

```python
merged_data = pd.DataFrame()

for file in dta_files:
    data = pd.read_stata(directory + '\\' + file)
    merged_data = pd.concat([merged_data, data], join = 'outer', ignore_index=True)
```

```python
merged_data
```

```python
merged_data.to_csv('DATA\china_industry_1998-2013.csv', index=False)
```
