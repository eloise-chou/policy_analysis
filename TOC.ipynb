# 介紹

此專案將討論北京-天津-河北地區發展計劃是否有效。該計劃於2007年實施，主要內容如下：

1. 合作生產和銷售煤炭、天然氣、電力、石油和其他能源資源，以及能源開發。
1. 將佔地面積較大的行業集中在河北地區，因為北京和天津的土地資源短缺。
1. 發展技術密集型產業和現代服務業，轉型傳統產業。

中國政府常常通過政策來干預產業發展，如果這些政策被證明是有效的，其他國家可以從這個發展計劃中學習，找到適合在其國家實施的計劃。此專案中使用的數據是 1998 年至 2013 年的中國工業數據，並將使用 DID 方法來估計該計劃對廠商收益的因果影響。

# 結論

隨著控制的變數越來越多，計劃對利潤的影響有所減弱，但仍然顯著。另外在早期階段並沒有此計劃對廠商收益有影響的證據。這可能是因為行也升級和遷移工廠需要時間及成本，因此計劃的影響只在後期階段才會顯著。

然而，此專案使用的數據存在一些限制。由於中國每年都有許多政策在實施，且數據的時間跨度很大，因此無法確定結論是否受其他政策的影響。另外此數據是每年對廠商進行問卷調查收集的，並不是追蹤數據，這一點也會對分析造成一定的影響。
